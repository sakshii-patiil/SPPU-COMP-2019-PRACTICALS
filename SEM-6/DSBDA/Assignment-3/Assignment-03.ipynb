```python
import pandas as pd
df = pd.read_csv('iris.csv')
print(df)
```

```python
setosa = df[df['variety'] == "Setosa"]
print("Setosa dataset\n\n")
print(setosa)
print("\n\nMin\t    ", setosa["sepal.length"].min(),"\t ", setosa["sepal.width"].min(),"\t       ", setosa["petal.length"].min(),"\t    ", setosa["petal.width"].min())
print("\n\nMax\t    ", setosa["sepal.length"].max(),"\t ", setosa["sepal.width"].max(),"\t       ", setosa["petal.length"].max(),"\t    ", setosa["petal.width"].max())
print("\n\nMean\t    ", setosa["sepal.length"].mean(),"\t ", setosa["sepal.width"].mean(),"\t    ", setosa["petal.length"].mean(),"\t  ", setosa["petal.width"].mean())
print("\n\nMedian\t    ", setosa["sepal.length"].median(),"\t ", setosa["sepal.width"].median(),"\t       ", setosa["petal.length"].median(),"\t    ", setosa["petal.width"].median())
print("\n\nStdDeviation", setosa["sepal.length"].std(),"  ", setosa["sepal.width"].std(),"  ", setosa["petal.length"].std(),"\t    ", setosa["petal.width"].std())
print("\n\nQuantile-1\t    ", setosa["sepal.length"].quantile(0.1),"\t ", setosa["sepal.width"].quantile(0.1),"\t", setosa["petal.length"].quantile(0.1)," ", setosa["petal.width"].quantile(0.1))
print("\n\nQuantile-2\t    ", setosa["sepal.length"].quantile(0.5),"\t  ", setosa["sepal.width"].quantile(0.5),"\t ", setosa["petal.length"].quantile(0.5),"\t   ", setosa["petal.width"].quantile(0.5))
print("\n\nQuantile-3\t    ", setosa["sepal.length"].quantile(0.9),"\t  ", setosa["sepal.width"].quantile(0.9),"\t ", setosa["petal.length"].quantile(0.9),"\t   ", setosa["petal.width"].quantile(0.9))
```
